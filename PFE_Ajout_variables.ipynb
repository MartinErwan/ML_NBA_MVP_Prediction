```python
import pandas as pd
import sqlite3
pd.options.display.max_columns = 205
import warnings
warnings.filterwarnings("ignore")
```

```python
FinalStatcon = sqlite3.connect('finalstat.sqlite')
```

```python
FinalStat = pd.read_sql("""SELECT * from FinalStat """, FinalStatcon)
```

```python
DF = FinalStat.copy()
```

```python
#Récupérations des index des MVPS de chaque année
index_mvps = DF.sort_values('Share',ascending=False).drop_duplicates('Year',keep='first').index 
DF[DF.index.isin(index_mvps)]
del DF["level_0"]
```

## Création des colonnes "MVP l'année précédente" et "MVP deux années avant"

Maintenant que nous avons récupérer tous les MVPS, on peut maintenant créer nos 2 colonnes. Pour se faire, nous allons travailler sur un sous ensemble de nos données qui contient uniquement les informations qui nous intéressent dans cette partie.

```python
temp = DF.loc[(DF['Year'] != 2023) & (DF.index.isin(index_mvps)), ['Player', 'Year']].sort_values('Year')
temp["MVP"] = 1
temp
```

```python
DF = DF.join(temp["MVP"])
DF["MVP"] = DF["MVP"].fillna(0)
DF[DF.index.isin(index_mvps)].head(10)
```

```python
DF_bis = DF[DF["Year"]<2022].groupby(["Player"]).sum()
DF_bis[DF_bis.index == "Karl Malone"]
```

```python
DF.sort_values(["Player","Year"],inplace = True)
DF["HAVE_2+_MVPS"] = 0
Year = 2022
years = DF["Year"].unique()[1:]
for year in years :
    DF_bis = DF[DF["Year"]<year].groupby(["Player"]).sum()
    
    for player in DF_bis.index:
        if DF_bis.loc[player,"MVP"]>=2:
            DF.loc[(DF["Player"]==player) & (DF["Year"]==year), "HAVE_2+_MVPS"] = 1


```

```python
DF[DF["Player"] == "Giannis Antetokounmpo"]
```

```python
DF.sort_values(["Player","Year"],inplace = True)
DF["MVP_Y-1"] = 0 #A été MVP l'année d'avant
DF["MVP_Y-1_Y-2"] = 0#A été MVP les deux années précédentes
for i in range(2,len(DF)):
    if(DF["Player"].iloc[i]==DF["Player"].iloc[i-1] and DF["MVP"].iloc[i-1]==1):
        DF["MVP_Y-1"].iloc[i] = 1
    if(DF["Player"].iloc[i]==DF["Player"].iloc[i-1]==DF["Player"].iloc[i-2] and DF["MVP"].iloc[i-1]==DF["MVP"].iloc[i-2]==1):
        DF["MVP_Y-1_Y-2"].iloc[i] = 1
    
    
    
```

```python
del DF["MVP"]
```

Enfin, on met à jour la base de données pour garder ces changements.

```python
DF.to_sql("FinalStat", con = FinalStatcon,if_exists = "replace")
```

```python
FinalStatcon.close()
```

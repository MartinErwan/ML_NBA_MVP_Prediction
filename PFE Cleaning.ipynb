# PFE : Analyse de données statistiques sportives

## II. Cleaning des données

Comme vous le savez, avant d'effectuer une prédiction sur une variable que l'on souhaite prédire, une des phases fondamentale d'un Data Scientist est le Cleaning.
Nous allons par conséquent effectuer un nombre de modification sur notre Dataset pour optimiser la prédiciton de nos algorithmes.
Le but ici serait de clean les datasets :
- MVPS 
- PLAYERS
- TEAMS

Afin faire un merge final et enfin pouvoir faire nos prediction sur un même et unique fichier

## CODE

### Import des librairies nécessaires

```python
import pandas as pd
import sqlite3
import re
```

### Lecture des fichiers SQLite

```python
path = ""
databaseMVP = path + 'mvp.sqlite'
MVPcon = sqlite3.connect(databaseMVP)
```

```python
MVPS = pd.read_sql("""SELECT * from MVP""", MVPcon)
```

```python
path = ""
databasePLAYERS = path + 'players.sqlite'
PLAYERScon = sqlite3.connect(databasePLAYERS)
```

```python
PLAYERS = pd.read_sql("""SELECT * from PLAYERS""", PLAYERScon)
```

```python
path = ""
databaseTEAMS = path + 'teams.sqlite'
TEAMScon = sqlite3.connect(databaseTEAMS)
```

```python
TEAMS = pd.read_sql("""SELECT * from TEAMS""", TEAMScon)
```

### Clean du dataset MVPS

```python
MVPS
```

#### Nous pouvons voir ici que les colonnes sur les statistiques ne servent pas à grande chose en raison de leur présence dans le dataset player, nous allons par conséquent garder uniquement les colonnes necessaire à ce dataset

```python
MVPS = MVPS[["Player", "Year", "Pts Won", "Pts Max", "Share"]]
```

```python
MVPS
```

### Clean du dataset PLAYERS

```python
PLAYERS
```

#### Nous pouvons voir ici que les colonnes "index", "Rk" ne servent à rien à l'intérêt du dataset qui est censé nous répertorier les statistiques de chaque joueur sur une année

```python
PLAYERS.drop(columns = ["index","Rk"], inplace = True)
```

```python
PLAYERS
```

#### Un autre soucis qui peut être posé quand nous allons merge les datasets, serait la présence d'astérisque à la fin des joueurs qui ne sont pas présentes dans les autres dataset.

```python
PLAYERS["Player"].head(50)
```

#### L'idée serait donc tout simplement de les enlever pour nous faciliter le merge

```python
PLAYERS["Player"] = PLAYERS["Player"].str.replace("*", "", regex = False)
```

```python
PLAYERS.head(20)
```

#### Lors de la lecture du dataframe PLAYERS nous pouvons remarquer que certains joueurs ont plusieurs lignes dans le dataframe sur une même année qui est dû au faite que ces joueurs ont joué dans plusieurs équipes différentes lors d'une même saison

#### L'idée ici serait uniquement de garder le statistiques globales sur l'année, donc nous allons nous intéresser à la valeur "Tm" = "TOT" qui correspond à la statistique globale d'un joueur sur une année 

```python
PLAYERS.loc[PLAYERS["Player"] == "George Johnson"]
```

```python
def uniqueROW(df):
    if df.shape[0] == 1:
        return df
    else:
        row = df[df["Tm"] == "TOT"]
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row

PLAYERS = PLAYERS.groupby(["Player", "Year", "Pos"]).apply(uniqueROW)
```

```python
PLAYERS
```

#### Les deux première colonnes ont été crée par principe lors du merge, nous pouvons les drops sans problème en raison de leur présence

```python
PLAYERS.index = PLAYERS.index.droplevel()
```

```python
PLAYERS.index = PLAYERS.index.droplevel()
```

```python
PLAYERS.index = PLAYERS.index.droplevel()
```

```python
PLAYERS
```

```python
PLAYERS.loc[PLAYERS["Player"] == "George Johnson"]
```

#### Maintenant que nos deux dataframes sont prêts nous pouvons déjà merge ces deux là en selectionnant une méthode 'outer' centré sur les colonnes 'Player' et 'Year'

```python
combined = PLAYERS.merge(MVPS, how = "outer", on = ["Player", "Year"])
```

```python
MVPS[MVPS["Year"]==2023]
```

```python
combined.loc[combined["Player"] == "George Johnson"]
```

#### Ici, le problème chronique des données trop ancienne, est l'absence des colonnes "Pts Won", "Pts Max" et "Share", nous pouvons les remplacer directement par 0 qui n'a pas d'impacte sur la prédiction

```python
combined[["Pts Won", "Pts Max", "Share"]] = combined[["Pts Won", "Pts Max", "Share"]].fillna(0)
```

```python
TEAMS.head(30)
```

#### Lors du scrapping du dataframe TEAMS, nous avons effectué un scrap total qui a aussi exporter les titres de chaque section, à savoir les divisions qui sont inutiles à notre future prédiction, nous pouvons donc les drop

```python
TEAMS = TEAMS[~TEAMS["W"].str.contains("Division")]
```

```python
TEAMS.head(30)
```

#### Même problème qu'avant concernant les astérisques, nous pouvons les enlever de la même manière

```python
TEAMS["Team"] = TEAMS["Team"].str.replace("*", "", regex = False)
```

ETAPE EN PLUS POUR L'ANNEE EN COURS

En effet, dans l'année en cours, on remarque que les noms des équipes sont suivis d'un chiffre en parenthèse

```python

TEAMS['Team'] = TEAMS['Team'].apply(lambda x: re.sub(r'\(\d+\)', '', x).strip())
```

```python
TEAMS.head(30)
```

```python
TEAMS["Team"].unique()
```

```python
combined["Tm"].unique()
```

#### Nous rentrons dans la phase finale du cleaning, donc le merge final avec le dataframe TEAMS, cependant, étant donné que nous le centrant sur "Players" et "Years", nous devons effectuer des modifications sur le nom des équipes dans notre dataset précédent qui note les teams par abbréviation et non par nom complet.

#### Nous avons crées un fichier nicknames.txt avec comme contenu 'Abbreviation' = 'Nom' sous forme de dictionnaire pour remplacer chaque abbréviation par leur nom complet pour un merge efficace.

```python
nicknames = {}

with open("nicknames.txt") as f:
    lines = f.readlines()
    for line in lines[1:]:
        abbrev, name = line.replace("\n", "").split(",")
        nicknames[abbrev] = name
```

```python
combined["Team"] = combined["Tm"].map(nicknames)
```

```python
combined["Team"].unique()
```

#### Nous pouvons enfin concater nos 2 derniers dataframes avec le dataframe TEAMS fraichement préparé

```python
FinalWorkDF = combined.merge(TEAMS, how = "inner", on = ["Team", "Year"])
```

```python
FinalWorkDF
```

```python
FinalWorkDF.dtypes
```

#### Certaines statistiques restent en "object" qui peuvent poser un problème, nous allons donc les convertir en extension numérique tout en laissant les variables s'apparentant à un string en object comme par exemple 'Player' via le paramètre " errors = 'ignore' "

```python
FinalWorkDF = FinalWorkDF.apply(pd.to_numeric, errors = "ignore")
```

```python
FinalWorkDF.dtypes
```

```python
FinalWorkDF["GB"].unique()
```

#### Nous pouvons voir que la statistique "GB" qui correspond à "Games Back" reste en "object", après inspection nous pouvons voir que ceci est dû au faite qu'il y a une valeur '—' qui veut tout simplement dire que l'équipe en question est en tête, donc nous pouvons remplacer cette valeur par '0.0'

```python
FinalWorkDF["GB"] = FinalWorkDF["GB"].str.replace("—", "0.0")
```

```python
FinalWorkDF["GB"].unique()
```

```python
FinalWorkDF["GB"] = pd.to_numeric(FinalWorkDF["GB"])
```

```python
FinalWorkDF["GB"].unique()
```

#### Inspectons les valeurs nulles qui peuvent poser un problème lors de la prédiction (A faire)

```python
pd.isna(FinalWorkDF).sum()
```

```python
FinalWorkDF[pd.isna(FinalWorkDF["GS"])]["Year"].unique()
```

#### Nous pouvons voir que la statistiques "GS" qui correspond à la titularisation d'un joueur est manquante sur les années 1980 et 1981, qui reste quand même une statistiques importantes à la prédiction du MVP, par conséquent nous pouvons drop ces deux années précises.

```python
FinalWorkDF.drop(FinalWorkDF.loc[FinalWorkDF["Year"] == 1980].index, inplace = True)
```

```python
FinalWorkDF.drop(FinalWorkDF.loc[FinalWorkDF["Year"] == 1981].index, inplace = True)
```

```python
pd.isna(FinalWorkDF).sum()
```

#### Nous pouvons noter que quelques statistiques restent nulles, cependant c'est facilement expliquer par leurs calcule
#### En effet le calcule pour la colonne 3P% est calculé par le ratio entre 3P (3 points goal) et 3PA (3 points attempt), donc si le joueur n'a jamais tenté un tire à 3 points sa colonne sera une division par 0 donc remplacé par un NA

```python
FinalWorkDF[pd.isnull(FinalWorkDF["3P%"])][["Player", "3PA"]]
```

#### Nous pouvons remplacer ces valeurs manquantes par 0, ce qui n'est pas totalement vraie mais nous pouvons quand même remplacer car ca n'influe pas sur la prédiction finale qui est de prédire le MVP, et le MVP sera un joueur ayant marqué des buts à 3 points

#### Nous pouvons voir le même constat pour les autres colonnes comme "FT%" et "FG%" qui se calcule de la meme manière avec une division de deux colonnes

```python
FinalWorkDF = FinalWorkDF.fillna(0)
```

```python
pd.isna(FinalWorkDF).sum()
```

```python
FinalStat = sqlite3.connect('finalstat.sqlite')
```

```python
FinalWorkDF.to_sql("FinalStat", con = FinalStat,if_exists="replace")
```
